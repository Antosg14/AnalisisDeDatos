```python
import pandas as pd
tabla = 'casas.csv'
df = pd.read_csv(tabla)

print(df)
```

```python
# CAMBIAR TITULOS DE COLUMNA
nombres_columnas = {
    'TÍTULO': 'Titulo',
    'TIPO DE PROPIEDAD': 'Tipo_de_propiedad',
    'PRECIO': 'Precio',
    'RECÁMARAS': 'Recamaras',
    'BAÑOS': 'Lavado',
    'TAMAÑO DE PROPIEDAD': 'M2',
    'FECHA DE DISPONIBILIDAD': 'Disponibilidad'
}
df = df.rename(columns=nombres_columnas)

print(df)
```

```python

# Convertir las columnas a tipo de dato entero
df['Precio'] = df['Precio'].astype(int)
df['Recamaras'] = df['Recamaras'].astype(int)
df['Lavado'] = df['Lavado'].astype(int)
df['M2'] = df['M2'].astype(int)
```

```python
# Calculamos la correlacion entre el precio y el tamaño de la propiedad

import matplotlib.pyplot as plt
correlacion = df['Precio'].corr(df['M2'])
print("Correlación entre Precio y Tamaño de propiedad:", correlacion)

# Crear un gráfico de dispersión
plt.scatter(df['M2'], df['Precio'])
plt.xlabel('Tamaño de propiedad (M2)')
plt.ylabel('Precio')
plt.title('Relación entre Precio y Tamaño de propiedad')
plt.show()
```

```python
correlacion1 = df['Precio'].corr(df['Recamaras'])
print("Correlación entre Precio y Recamaras de la propiedad:", correlacion)

# Crear un gráfico de dispersión
plt.scatter(df['Recamaras'], df['Precio'])
plt.xlabel('Recamaras de la propiedad')
plt.ylabel('Precio')
plt.title('Relación entre Precio y Recamaras de la propiedad')
plt.show()
```

```python
correlacion = df['Precio'].corr(df['Lavado'])
print("Correlación entre Precio y Lavado(baños) de la propiedad:", correlacion)

# Crear un gráfico de dispersión
plt.scatter(df['Lavado'], df['Precio'])
plt.xlabel('Lavado')
plt.ylabel('Precio')
plt.title('Relación entre Precio y Lavado')
plt.show()
```

```python
correlacion = df['Precio'].corr(df['M2'])
print("Correlación entre Precio y Tamaño de propiedad:", correlacion)

# Crear un gráfico de dispersión
plt.scatter(df['M2'], df['Precio'])
plt.xlabel('Tamaño de propiedad (M2)')
plt.ylabel('Precio')
plt.title('Relación entre Precio y Tamaño de propiedad')
plt.show()
```

```python
correlacion = df['Lavado'].corr(df['Recamaras'])
print("Correlación entre Lavado y Recamaras:", correlacion)

# Crear un gráfico de dispersión
plt.scatter(df['Lavado'], df['Recamaras'])
plt.xlabel('Lavado')
plt.ylabel('Recamaras')
plt.title('Relación entre Lavado y Recamaras')
plt.show()
```

```python
from sklearn.cluster import KMeans

X = df[['Recamaras', 'Lavado', 'M2']]
kmeans = KMeans(n_clusters=3)  # Especifica el número de clusters deseado
kmeans.fit(X)
df['Cluster'] = kmeans.labels_

# Visualización de clusters
plt.scatter(df['M2'], df['Precio'], c=df['Cluster'])
plt.xlabel('M2')
plt.ylabel('Precio')
plt.title('Clusters de propiedades')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar las columnas de interés
variables = ['Precio', 'Recamaras', 'M2', 'Lavado']
data = df[variables]

# Calcular la matriz de correlación
correlation_matrix = data.corr()

# Crear el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de calor de correlación')
plt.show()
```
